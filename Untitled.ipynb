```python
total_per_kategori = df.groupby('major_category')['value'].sum()

# Membuat bar chart
plt.figure(figsize=(10, 6))
total_per_kategori.plot(kind='bar', color='green', alpha=0.7)
plt.xlabel('Kategori Utama')
plt.ylabel('Total Nilai')
plt.title('Bar Chart Total Nilai per Kategori Utama')
plt.xticks(rotation=45)  
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
total_per_kategori = df.groupby('major_category')['value'].sum()


plt.figure(figsize=(8, 8))
plt.pie(total_per_kategori, labels=total_per_kategori.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  
plt.title('Pie Chart Total Nilai per Kategori Utama')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot([df[df['year'] == tahun]['value'] for tahun in df['year'].unique()],
            labels=df['year'].unique())
plt.xlabel('Tahun')
plt.ylabel('Nilai')
plt.title('Box Plot Nilai per Tahun')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['year'], df['value'], alpha=0.5)
plt.xlabel('Tahun')
plt.ylabel('Nilai')
plt.title('Scatter Plot Nilai per Tahun')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
